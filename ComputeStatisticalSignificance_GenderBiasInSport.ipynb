Given the results for perplexity calculations in the questions asked to male athletes and question asked to female athletes datasets, we want to calculate a statistical significance of the results. We will compute these in two ways, a t-test and a Mann Whitney U-test, representing the relaxation of the normality assumption and the no-relaxation scenario.

```python
from scipy import stats
```

```python
import pandas as pd
from scipy.stats import mannwhitneyu
```

```python
import numpy as np
```

```python
perplexity_df = pd.read_csv('/content/questions_perplexity.xlsx - questions_perplexity.csv')
```

```python
perplexity_df.head()
```

First, we run a t-test

```python
# Assuming perplexity_df is your DataFrame with 'gender' and 'perplexity' columns
# Replace this with your actual DataFrame
# perplexity_df = ...

# Extract perplexity scores for male and female populations
M_perplexity = perplexity_df[perplexity_df['gender'] == 'M']['perplexity'].values
F_perplexity = perplexity_df[perplexity_df['gender'] == 'F']['perplexity'].values

# Perform Mann-Whitney U test
statistic, p_value = mannwhitneyu(M_perplexity, F_perplexity, alternative='two-sided')

# Print the results
print("U-statistic:", statistic)
print("P-value:", p_value)

# Check for statistical significance
alpha = 0.05
if p_value < alpha:
    print("The difference in distributions is statistically significant.")
else:
    print("There is no statistically significant difference in distributions.")
```
